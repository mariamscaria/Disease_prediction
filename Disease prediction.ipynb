```python
#--- Import matplotlib and  seaborn---
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/nimikaringada/downloads/heart_cleveland_upload.csv')

# Define numeric features
numeric_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'condition']

# Calculate the correlation between numeric features
corr_data = df[numeric_features].corr()

# Create a heatmap to visualize the correlation
plt.figure(figsize=(10, 8))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
